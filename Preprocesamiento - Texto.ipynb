# Preprocesamiento de texto

En este notebook explicamos y aplicamos algunas de las técnicas más comunes en el preprocesamiento de texto

```python
texto = """
La sociedad francesa estaba dividida en estamentos dependiendo de sus clases sociales, el poder mas alto lo tenía el rey, detrás estaban la nobleza y el clero y el nivel mas bajo de poder lo tenia el tercer estado que estaba constituido por la burguesía, los artesanos y los campesinos.

Los Estados Generales eran una asamblea, compuesta por tres ordenes separados: el clero, la nobleza y el grupo formado por burguesía y campesinado. Este último orden se conoce como el tercer estadeo, término que usaremos para referirnos a él en lo sucesivo. Dicha asamblea se había citado por ultima vez en 1614 y el dramatismo de la situación obligó al gobierno a convocarla nuevamente.

Luis XVI cedió a las presiones de la reina María Antonieta y del conde de Artois y dio instrucciones para que varios regimientos extranjeros leales se concentraran en París y Versailles. Al mismo tiempo, Necker fue nuevamente destituido. El pueblo de París respondió con la insurrección ante estos actos de provocación; los disturbios comenzaron el 12 de julio, y las multitudes asaltaron y tomaron La Bastilla -una prisión real que simbolizaba el despotismo de los Borbones- el 14 de julio.

El 5 de octubre de 1789, las mujeres parisinas partieron desde los barrios obreros hacia la residencia real de Versailles, este suceso dió comienzo a la revolución.

A fines de 1792 comenzó el proceso de Convención contra Luis XVI, quien fue juzgado y condenado a la guillotina por mayoría de votos. El 21 de enero de 1793, Luis subió al cadalso, inconmovible hasta el último momento en el sentimiento de su inocencia. La noticia de la muerte del rey produjo indignación en Inglaterra, la que despidió al embajador o representante francés. Francia contestó declarando la guerra a Inglaterra y a Holanda, su aliada.
"""

print(texto)
```

## Eliminación de signos de puntuación.

```python
import string
string.punctuation

for punctuation in string.punctuation:
    texto = texto.replace(punctuation, '')

print(texto)
```

## Eliminación de números


```python
import re
texto = re.sub(r'\d+', '', texto)
print(texto)
```

## Normalización a minúsculas


```python
texto = texto.lower()
print(texto)
```

## Eliminación de stopwords en Español

```python
texto = texto.replace("\n", ' ')
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

texto = ' '.join([word for word in texto.split(" ") if word not in (stopwords.words('spanish'))])
print(texto)
```

## Normalización de tildes

```python
texto = texto.replace('á', 'a')
texto = texto.replace('é', 'e')
texto = texto.replace('í', 'i')
texto = texto.replace('ó', 'o')
texto = texto.replace('ú', 'u')
print(texto)
```

```python
texto = ' '.join(texto.split())
```

## Creación de bigramas

```python
from nltk import ngrams
from collections import Counter

n = 2
bigrams = ngrams(texto.split(), n)
most_common_bigrams = Counter(bigrams).most_common()
most_common_bigrams = [bigram for bigram in most_common_bigrams if bigram[1] > 1]
print(most_common_bigrams)

# Sustituir bigramas en texto
for bigram in most_common_bigrams:
    texto = texto.replace(f"{bigram[0][0]} {bigram[0][1]}", f"{bigram[0][0]}_{bigram[0][1]}")
print(texto)
```

## Tokenización

```python
tokens = texto.split()
```

```python
tokens[0:10]
```

## Stemming


```python
from nltk.stem import SnowballStemmer
stemmer = SnowballStemmer('spanish')

stemmer.stem("coche")
```

```python
from nltk.stem import SnowballStemmer
stemmer = SnowballStemmer('spanish')
stemmed_tokens = [stemmer.stem(token) for token in tokens]
print(stemmed_tokens)
```
